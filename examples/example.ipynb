# Persistence1D.jl
Julia interface to the extrema finding algorithm [Persistence1D](https://github.com/yeara/Persistence1D.git) (public domain).

## Setup

```julia
using Persistence1D
using PyPlot
PyPlot.svg(true)
```

```julia
x = 1:600; 
SineLowFreq = sin(x * 0.01 * pi);
SineMedFreq = 0.25 * sin(x * 0.01 * pi * 4.9);
SineHighFreq = 0.15 * sin(x * 0.01 * pi * 12.1);
y = SineLowFreq + SineMedFreq + SineHighFreq;
nothing
```

## Find extrema

Find all extrema:

```julia
valleys,peaks = find_extrema(y)
```

Find all extrema with persistence $>0.5$:

```julia
valleys2,peaks2 = find_extrema(y,0.5)
```

```julia
plot(x,y)
plot(x[valleys2],y[valleys2], "o")
plot(x[peaks2],y[peaks2], "o")
plot(x[valleys],y[valleys], ".")
plot(x[peaks],y[peaks], ".")
```

## Reconstruct persistent features

```julia
plot(x,y)
plot(x,reconstruct_persistent(y,0.5))
```
